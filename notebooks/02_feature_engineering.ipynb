```python
import pandas as pd
 
df = pd.read_csv('../data/raw/hotel_bookings.csv')

df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'] + '-' +
    df['arrival_date_day_of_month'].astype(str)
)
daily = (
    df.groupby('arrival_date')
      .size()
      .reset_index(name='bookings')
)

daily = daily.sort_values('arrival_date').reset_index(drop=True)
daily.head()
```

```python
# create time based features
daily['day_of_week'] = daily['arrival_date'].dt.dayofweek
daily['is_weekend'] = daily['day_of_week'].isin([5,6]).astype(int)
daily['month'] = daily['arrival_date'].dt.month
```

```python
daily.head(5)
```

```python
daily['lag_1'] = daily['bookings'].shift(1)
daily['lag_7'] = daily['bookings'].shift(7)


```

```python
daily['rolling_7'] = daily['bookings'].rolling(window=7).mean()
daily['rolling_14'] = daily['bookings'].rolling(window=14).mean()
```

```python
daily.head(20)
```

```python
daily.isnull().sum()
```

```python
daily_clean = daily.dropna().reset_index(drop=True)
```

```python
daily_clean.isnull().sum()
```

```python
features = [
    'day_of_week',
    'is_weekend',
    'month',
    'lag_1',
    'lag_7',
    'rolling_7',
    'rolling_14'
]

target = 'bookings'
X = daily_clean[features]
y = daily_clean[target] 
```

```python
daily_to_save = daily[['arrival_date', 'bookings']]
daily_to_save.to_csv("../data/processed/daily_demand.csv", index=False)

print("Processed demand data saved")
```
